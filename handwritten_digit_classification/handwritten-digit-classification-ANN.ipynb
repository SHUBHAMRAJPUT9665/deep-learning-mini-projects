```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
tfData = keras.datasets.mnist.load_data()
```

```python
tfData
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
from tensorflow.keras.layers import Dropout

model = Sequential()
model.add(keras.Input(shape=(28,28)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.3)) # Added Dropout layer
model.add(Dense(10,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam')
print("Model redefined and recompiled with an additional Dense layer and a Dropout layer.")
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'],color='red',label='train')
plt.plot(history.history['val_loss'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.show()
```

# Task
Apply L2 kernel regularization with a strength of 0.001 to the Dense layers of the existing neural network model, recompile the model with the same loss, optimizer, and metrics, then train and evaluate the regularized model on the MNIST dataset, and finally summarize the effects of L2 regularization on the model's performance and overfitting behavior.

## Apply L2 Regularization

### Subtask:
Modify the existing Dense layers in the model to include L2 kernel regularization to reduce overfitting. I will set a regularization strength (e.g., 0.001) for this experiment.


**Reasoning**:
The subtask requires modifying the existing Dense layers to include L2 kernel regularization. I will redefine the model, incorporating the l2 regularizer in each Dense layer, while keeping the rest of the model architecture and compilation parameters similar to the previous version.



```python
from tensorflow.keras.regularizers import l2

model = Sequential()
model.add(keras.Input(shape=(28,28)))
model.add(Flatten())
model.add(Dense(128,activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dense(128,activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dropout(0.3))
model.add(Dense(10,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
print("Model redefined and recompiled with L2 regularization on Dense layers.")
```

**Reasoning**:
Now that the model has been redefined with L2 regularization and recompiled, the next step is to train it using the `X_train` and `y_train` data, similar to the previous training, and store the training history for later analysis.



```python
history_regularized = model.fit(X_train,y_train,epochs=10,validation_split=0.2)
print("Regularized model training complete.")
```

```python
y_prob_regularized = model.predict(X_test)
y_pred_regularized = y_prob_regularized.argmax(axis=1)
print("Predictions generated for the regularized model.")
```

**Reasoning**:
Now that predictions for the regularized model (`y_pred_regularized`) have been generated, the next logical step is to calculate the accuracy of this model on the test set (`X_test`) by comparing `y_pred_regularized` with the actual test labels (`y_test`). I will use `sklearn.metrics.accuracy_score` for this.



```python
from sklearn.metrics import accuracy_score
accuracy_regularized = accuracy_score(y_test, y_pred_regularized)
print(f"Accuracy of the regularized model: {accuracy_regularized}")
```

**Reasoning**:
To compare the performance of the regularized model with the previous model, I will visualize the training and validation loss and accuracy curves for both models side-by-side. This will highlight the impact of L2 regularization on the model's behavior regarding overfitting.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

# Plotting Loss for both models
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], color='red', label='Train Loss (Original)')
plt.plot(history.history['val_loss'], color='blue', label='Validation Loss (Original)')
plt.plot(history_regularized.history['loss'], color='red', linestyle='--', label='Train Loss (Regularized)')
plt.plot(history_regularized.history['val_loss'], color='blue', linestyle='--', label='Validation Loss (Regularized)')
plt.title('Model Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plotting Accuracy for both models
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], color='green', label='Train Accuracy (Original)')
plt.plot(history.history['val_accuracy'], color='purple', label='Validation Accuracy (Original)')
plt.plot(history_regularized.history['accuracy'], color='green', linestyle='--', label='Train Accuracy (Regularized)')
plt.plot(history_regularized.history['val_accuracy'], color='purple', linestyle='--', label='Validation Accuracy (Regularized)')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
print("Loss and accuracy plots for both models displayed for comparison.")
```
